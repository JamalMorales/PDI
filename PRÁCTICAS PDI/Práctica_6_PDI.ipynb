```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.optimize import minimize

#Jamal Morales Sánchez 2064497
#Evelyn Rubí Martinez Ruiz 2071292
#Abigail Montserrat Nieto Solís 2059067
#Reynaldo Holsen Gallegos González	1963727
#Jesús Antonio Segura Vences	1851396

import numpy as np
import matplotlib.pyplot as plt
import cv2

#Importación de imágenes
img1 = cv2.imread('IMAGENES/Tomografía uno.jpg', 0)
img2 = cv2.imread('IMAGENES/Tomografía dos.jpg', 0)

#Reescalado de imágenes
imgs1 = cv2.resize(img1, (256, 256))
imgs2 = cv2.resize(img2, (256, 256))

#Sumas
suma = cv2.addWeighted(imgs1, 0.8, imgs2, 0.2, 0)
spro = cv2.add(imgs1, imgs2)

#Restas
resta = cv2.absdiff(imgs1, imgs2)

#Multiplicaciones
mult2_1 = imgs2 * 3
mult2_2 = imgs2 * 1
mult2_3 = imgs2 * 6

mult1_1 = imgs1 * 5
mult1_2 = imgs1 * 1.5
mult1_3 = imgs1 * 10

#Multiplicación elemento x elemento
mexe = imgs1 * imgs2

plt.figure(figsize=(12,4))
plt.subplot(3,4,1)
plt.imshow(imgs1, cmap = 'gray')
plt.title('Original 1')
plt.axis('off')

plt.subplot(3,4,2)
plt.imshow(imgs2, cmap = 'gray')
plt.title('Original 2')
plt.axis('off')

plt.subplot(3,4,3)
plt.imshow(suma, cmap = 'gray')
plt.title('Suma Ponderada')
plt.axis('off')

plt.subplot(3,4,4)
plt.imshow(spro, cmap = 'gray')
plt.title('Suma de promedios')
plt.axis('off')

plt.subplot(3,4,5)
plt.imshow(resta, cmap='gray')
plt.title('Resta')
plt.axis('off')

plt.subplot(3,4,6)
plt.imshow(mult1_1, cmap='gray')
plt.title('Multiplicación x 5')
plt.axis('off')

plt.subplot(3,4,7)
plt.imshow(mult1_2, cmap='gray')
plt.title('Multiplicación x  1.5')
plt.axis('off')

plt.subplot(3,4,8)
plt.imshow(mult1_3, cmap='gray')
plt.title('Multiplicación x 10')
plt.axis('off')

plt.subplot(3,4,9)
plt.imshow(mult2_1, cmap='gray')
plt.title('Multiplicación x 3')
plt.axis('off')

plt.subplot(3,4,10)
plt.imshow(mult2_2, cmap='gray')
plt.title('Multiplicación x 1')
plt.axis('off')

plt.subplot(3,4,11)
plt.imshow(mult2_3, cmap='gray')
plt.title('Multiplicación x 6')
plt.axis('off')

plt.subplot(3,4,12)
plt.imshow(mexe, cmap='gray')
plt.title('Multiplicación elemento x elemtno')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

#Importación de imágenes
img1 = cv2.imread('IMAGENES/Tomografía uno.jpg', 0)

#Mejora de contraste
imge = cv2.equalizeHist(img1)

#Mostrar
plt.subplot(2,2,1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2,2,2)
plt.hist(img1.ravel(),256,[0,256])
plt.title('Histograma original')
plt.axis('on')

plt.subplot(2,2,3)
plt.imshow(imge, cmap='gray')
plt.title('Histograma ecualizado')
plt.axis('off')

plt.subplot(2,2,4)
plt.hist(imge.ravel(),256,[0,256])
plt.title('Histograma imagen ecualizada')
plt.axis('on')

plt.tight_layout()
plt.show()
```
